```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv('data/cleaned_jobs.csv').dropna()
```

```python
vec = CountVectorizer(max_features=500)
encoded_descriptions = vec.fit_transform(raw_documents=df['description'].values).toarray()

salary = df['salary_estimate'].values
```

```python
lr = LinearRegression()
lr.fit(encoded_descriptions, salary)
```

```python
sorted_words = np.array(vec.get_feature_names())[np.argsort(lr.coef_)][::-1]
for word, coef in zip(sorted_words, np.sort(lr.coef_)[::-1]):
    print(word, coef)
```
